# Covid data analysis

```python
import pandas as pd
import plotly.express as px
from plotly.figure_factory import create_table
```

```python
df = pd.read_csv("covid.csv")
df
```

```python
df.columns
```

```python
df = df[['Date', 'Country', 'Continent', 'Code', 'Cases',
       'Deaths']]
df["Date"] = pd.to_datetime(df["Date"], dayfirst=True)
```

```python
create_table(df.head())
```

```python
fig = px.pie(df, values='Cases', names='Continent', title="Covid_19 Cases Continent wise", hole=0.5)
fig.show()
```

```python
def create_pie(option):
    fig = px.pie(df, values=option, names='Continent', title="Covid_19 " + option + " Continent wise", hole=0.5)
    fig.show()
```

```python
create_pie("Deaths")
```

```python
nepal_data = df[df['Country'] == "Nepal"]
nepal_data
```

```python
fig = px.bar(nepal_data, x="Date", y="Cases")
fig.show()
```

```python
def create_country_bar(country, option):
    country_data = df[df['Country'] == country]
    fig = px.bar(country_data, x="Date", y=option, animation_frame="Date", animation_group="Country", color="Country")
    fig.show()
```

```python
create_country_bar("Iraq", "Deaths")
```
