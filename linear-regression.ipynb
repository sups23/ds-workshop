# Linear regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("USA_Housing.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
df = df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population', 'Price']]
df
```

```python
df.rename(columns={
    'Avg. Area Income': 'income',
    'Avg. Area House Age': 'house_age',
    'Avg. Area Number of Rooms': 'no_of_rooms',
    'Avg. Area Number of Bedrooms': 'no_of_bedrooms',
    'Area Population': 'population',
    'Price': 'price'
}, inplace=True)
df
```

```python
import seaborn as sns
```

```python
df.columns
```

```python
X = df[['income', 'house_age', 'no_of_rooms', 'no_of_bedrooms', 'population']]
Y = df[['price']]
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y, train_size=0.8, random_state=20)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
prediction = model.predict(x_test)
prediction
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, prediction)
```
