```python
import numpy as np
```

```python
days = np.arange(0,1000)
temp1 = np.random.randint(low=10, high=30, size=1000)
temp2 = np.random.randint(low=20, high=40, size=1000)
```

```python
plt.figure(figsize=(50,50))
plt.plot(days, temp1, color='r', marker='*')
plt.plot(days, temp2, color='g', marker='*')
plt.title("Chitwan Temperature", fontsize=20)
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.show()
```
